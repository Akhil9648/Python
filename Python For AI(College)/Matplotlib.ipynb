Create a line chart to analyze the sales

```python
plt.plot(x)
plt.xlabel('Index')
plt.ylabel('Sales')
plt.title('Line chart of Sales')
```

```python
sns.boxplot(df['Sales'])
```

Create a plot to compare average sales per category

```python
df
```

```python
plt.bar(df['Category'],df['Sales'])
plt.xlabel("Category")
plt.ylabel("Sales")
```

```python
df.groupby('Category')['Sales'].mean().plot(
    kind='bar',color='black',title='Average Sales for a Category'
)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

Create a plot to compare the distribution of data per category

```python
category_counts=df['Category'].value_counts()
category_counts.plot(
    kind='pie',autopct='%1.1f%%',startangle=90,title='Category Plot'
)
plt.ylabel(' ')
plt.show()
```

Create a Histogram for Profit Column

```python
plt.hist(df['Profit'])
plt.xlabel('Index')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
df['Profit'].plot(
    kind='hist',bins=20,color='green',alpha=0.7,title='Histogram of Profit'
)
plt.xlabel('Profit')
plt.show()
```

Create two subplots side by side to find the patterns of sales and profit column

```python
plt.subplot(2,2,1)
plt.plot(df['Profit'])
plt.title('Profit')
plt.subplot(2,2,2)
plt.plot(df['Sales'])
plt.title('Sales')
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,4))
axs[0].plot(df['Sales'],color='blue',linestyle='--')
axs[0].set_title('Sales')
axs[1].plot(df['Profit'],color='green',linestyle='--')
axs[1].set_title('Profit')
```

Create 4 subplots like 2X2 representation

```python
fig,axs=plt.subplots(2,2,figsize=(10,10))
axs[0,0].plot(df['Sales'],color='blue',linestyle='--')
axs[0,0].set_title('Sales')
axs[0,1].plot(df['Profit'],color='green',linestyle='--')
axs[0,1].set_title('Profit')
axs[1,0].plot(df['Sales'],color='blue',linestyle='--')
axs[1,0].set_title('Sales')
axs[1,1].plot(df['Profit'],color='green',linestyle='--')
axs[1,1].set_title('Profit')
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,10))
axs[0,0].plot(df['Sales'],color='orange')
axs[0,0].set_title('Line of Sales')
axs[0,1].hist(df['Sales'],color='green',bins=20)
axs[0,1].set_title('Histogram of Sales')
axs[1,0].plot(df['Profit'],color='orange')
axs[1,0].set_title('Line of Profit')
axs[1,1].hist(df['Profit'],color='green',bins=20)
axs[1,1].set_title('Histogram of Profit')
```
