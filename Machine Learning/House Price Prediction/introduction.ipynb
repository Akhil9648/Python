```python
def convert_sqft_to_num(x):
    try:
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        else:
            return float(x)
    except:
        return None
```

```python
data['total_sqft']= data['total_sqft'].apply(convert_sqft_to_num)  
```

```python
data['price_per_sqft'] = data['price']*100000/data['total_sqft']
```

```python
data['location']= data['location'].apply(lambda x: x.strip())
location_count= data['location'].value_counts()
```

```python
location_count_less_than_10= location_count[location_count<=10]
location_count_less_than_10
```

```python
data['location']=data['location'].apply(lambda x: 'other' if x in location_count_less_than_10 else x)
```

```python
data=data[((data['total_sqft']/data['bhk'])>=300)]
data.describe()
```

```python
def remove_outliers_sqft(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        st = np.std(subdf['price_per_sqft'])
        reduced_df = subdf[(subdf['price_per_sqft']>(m-st)) & (subdf['price_per_sqft']<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out
data = remove_outliers_sqft(data)
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df['price_per_sqft']),
                'std': np.std(bhk_df['price_per_sqft']),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df['price_per_sqft'] < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
data=bhk_outlier_remover(data)
```

```python
X=data.drop(columns=['price','size'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans=make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans, scaler, lr)
```

```python
for col in X_train.columns:
    print(col, X_train[col].dtype)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_prod_lr= pipe.predict(X_test)
```

```python
r2_score(y_test, y_prod_lr)
```

Lasso

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans, scaler, lasso)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_prod_lasso= pipe.predict(X_test)
```

```python
r2_score(y_test, y_prod_lasso)
```

Ridge

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_prod_ridge= pipe.predict(X_test)
```

```python
r2_score(y_test, y_prod_ridge)
```

```python
print("No Regularization:", r2_score(y_test, y_prod_lr))
print("Lasso Regularization:", r2_score(y_test, y_prod_lasso))
print("Ridge Regularization:", r2_score(y_test, y_prod_ridge))
```
