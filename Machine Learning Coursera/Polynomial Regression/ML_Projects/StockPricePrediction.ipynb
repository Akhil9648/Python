```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/Pranavd0828/NIFTY50-StockMarket/refs/heads/main/Dataset/NTPC.csv')
```

```python
data['Date']=pd.to_datetime(data['Date'])
data.sort_values('Date')
```

```python
data['Date_ordinal'] = data['Date'].map(lambda d: d.toordinal())
```

```python
X = data[['Date_ordinal']]
y = data['Close']
```

```python
lr=LinearRegression()
lr.fit(X,y)
```

```python
last_date=data['Date'].max()
future_dates=[last_date+pd.Timedelta(days=i) for i in range(1,6)]
future_ordinals=np.array([d.toordinal() for d in future_dates]).reshape(-1,1)
pred=lr.predict(future_ordinals)
```

```python
for date,price in zip(future_dates,pred):
    print(date.date(),round(price,2))
```
